```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Reach

```python
# algorithms' name
alg_name = ["DDPG", "DDPG_HER", "SAC_HER", "SAC_singleQ", "SAC", "TD3_HER", "TD3_singleQ", "TD3"]
```

```python
df_dic = {}
for name in alg_name:   
    df_dic[name] = pd.read_csv(f"Data/reach/success_rate/{name}.csv")
```

```python
df_dic.keys()
```

```python
color_map = {
    "DDPG": "#00A7A7",
    "DDPG_HER": "#66CCCC",

    "TD3": "#7B1FA2",
    "TD3_HER": "#A05EB5",
    "TD3_singleQ": "#C39ED6",

    "SAC": "#FF7043",
    "SAC_HER": "#FFA07A",
    "SAC_singleQ": "#FFD3B3"
}

plt.style.use("dark_background")
fig, ax = plt.subplots(figsize=(10, 6))

for name in alg_name:
    df = df_dic[name]
    ax.plot(df["Step"], df["Value"], label=name, color=color_map[name], linewidth=2)

ax.set_xlabel("Step", fontsize=12, fontname="Times New Roman")
ax.set_ylabel("Value", fontsize=12, fontname="Times New Roman")
ax.set_title("Success Rate Comparison on PandaReach-v3", fontsize=14, fontname="Times New Roman")
ax.grid(True, linestyle="--", alpha=0.3)
ax.legend(fontsize=10, loc="upper left")

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.tick_params(colors='white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.title.set_color('white')

plt.tight_layout()
plt.savefig("fig_reach")
plt.show()
```
